```python
import os
import asyncio
import datetime
import pandas as pd
from playwright.async_api import async_playwright
from unidecode import unidecode

# Carpeta de descargas
download_dir = os.path.join(os.getcwd(), "Descargas", str(year))
os.makedirs(download_dir, exist_ok=True)
print(f"📁 Carpeta de descargas creada en: {download_dir}")

# Datos usuario y URL
usuario = "rramos"
contrasena = "Matecaliente123"
url_reporte = (
    "https://movies.comscore.com/#reports/"
    "studio_reported/studio_reported_by_title?"
    "country_id=ALL&currency_type_no=1&gross_or_xtns=rev&"
    "pct_change_type=prev_week&srg_days_view=week_weekend"
)

peliculas = titulos_no_descargados_lista
estado_peliculas = []
```

```python
def normalizar(texto):
    return unidecode(texto).strip().lower()

async def ocultar_overlay(page):
    for _ in range(5):
        visible = await page.is_visible("div[data-testid='overlay']")
        if visible:
            print("Overlay visible, ocultándolo con JS...")
            await page.evaluate("""
                () => {
                    const ov = document.querySelector('div[data-testid="overlay"]');
                    if (ov) { ov.style.display = 'none'; }
                }
            """)
            await asyncio.sleep(1)
        else:
            break
    print("Overlay desapareció.")

async def cerrar_overlay_js(page):
    for intento in range(5):
        visible = await page.is_visible("div[data-testid='overlay']")
        if visible:
            print(f"Overlay visible (intento {intento+1}/5), ocultándolo con JS...")
            await page.evaluate("""
                () => {
                    const overlay = document.querySelector("div[data-testid='overlay']");
                    if (overlay) {
                        overlay.style.display = 'none';
                    }
                }
            """)
            await asyncio.sleep(1)
        else:
            print("Overlay desapareció.")
            break

async def forzar_reset_ui(page):
    try:
        await page.keyboard.press("Escape")
        await page.click("body")
        await cerrar_overlay_js(page)
        await asyncio.sleep(1)
    except:
        pass
```

```python
async def abrir_selector_pelicula(page, max_intentos=3):
    for intento in range(max_intentos):
        try:
            await ocultar_overlay(page)
            await asyncio.sleep(2)
            await page.wait_for_selector("button#mobile-film-picker", timeout=15000)
            await page.click("button#mobile-film-picker")
            await page.wait_for_selector("input[placeholder='Search Film']", timeout=15000)
            return True
        except Exception as e:
            print(f"❌ Intento {intento+1}/{max_intentos} fallo al abrir selector: {e}")
            await page.reload()
            await asyncio.sleep(5)
    return False

async def seleccionar_pelicula(page, nombre):
    if not await abrir_selector_pelicula(page):
        raise Exception("No se pudo mostrar el input para seleccionar película.")

    await cerrar_overlay_js(page)

    await page.wait_for_selector("button#mobile-film-picker", timeout=60000)
    btn_add = await page.query_selector("button#mobile-film-picker")
    await page.evaluate("el => el.scrollIntoView({block:'center'})", btn_add)
    await asyncio.sleep(0.3)
    await btn_add.click()
    await asyncio.sleep(1)

    try:
        await page.wait_for_selector("input[placeholder='Search Film']", timeout=60000)
    except Exception:
        raise Exception("No se pudo mostrar el input para seleccionar película.")

    input_film = await page.query_selector("input[placeholder='Search Film']")
    await page.evaluate("el => el.scrollIntoView({block:'center'})", input_film)
    await asyncio.sleep(0.3)
    await input_film.click()
    await input_film.fill("")
    await input_film.type(nombre)
    await asyncio.sleep(1)

    xpath_item = f"//ul[@role='listbox']//li[contains(translate(normalize-space(), 'ABCDEFGHIJKLMNOPQRSTUVWXYZ', 'abcdefghijklmnopqrstuvwxyz'), '{nombre.lower()}')]"

    try:
        await page.wait_for_selector(f"xpath={xpath_item}", timeout=60000)
        await page.click(f"xpath={xpath_item}")
    except:
        raise Exception(f"Película no disponible: {nombre}")

    try:
        await page.wait_for_selector("div[data-testid='overlay']", state="hidden", timeout=60000)
    except:
        await page.keyboard.press("Escape")
        await asyncio.sleep(1)

async def descargar_excel(page, nombre):
    await page.wait_for_selector("css=svg[data-icon='download']", timeout=60000)
    boton_export = await page.query_selector("css=svg[data-icon='download']")
    await page.evaluate("el => el.scrollIntoView({block:'center'})", boton_export)
    await boton_export.click()

    await page.wait_for_selector("xpath=//div[contains(text(),'Download Complete Report to Excel')]", timeout=60000)
    boton_xlsx = await page.query_selector("xpath=//div[contains(text(),'Download Complete Report to Excel')]")
    await page.evaluate("el => el.scrollIntoView({block:'center'})", boton_xlsx)

    async with page.expect_download(timeout=60000) as dl_info:
        await boton_xlsx.click()
    download = await dl_info.value

    suggested = download.suggested_filename
    _, ext = os.path.splitext(suggested)
    safe_name = nombre.replace(":", "_").replace(" ", "_") + f"_{year}"
    destino = os.path.join(download_dir, f"{safe_name}{ext}")
    await download.save_as(destino)
    print(f"✔️ Archivo guardado como: {destino}")
    return "✔️ Archivo guardado"

async def cerrar_exports(page):
    try:
        await page.wait_for_selector("css=li.gboe-nav-username.setting > div", timeout=6000)
        await page.click("css=li.gboe-nav-username.setting > div")
    except:
        await page.keyboard.press("Escape")
```

```python
async def run():
    browser_args = ['--no-sandbox']
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True, args=browser_args, slow_mo=100)
        context = await browser.new_context(accept_downloads=True, viewport={"width":1920, "height":1080})
        page = await context.new_page()

        # Login
        await page.goto("https://movies.comscore.com", wait_until="networkidle")
        await page.fill('css=input[placeholder="Username"]', usuario)
        await page.fill('css=input[placeholder="Password"]', contrasena)
        await page.keyboard.press("Enter")
        await asyncio.sleep(3)

        for i, peli in enumerate(peliculas, 1):
            print(f"\n===== ({i}/{len(peliculas)}) {peli} =====")
            try:
                await page.goto(url_reporte, wait_until="networkidle")

                try:
                    await page.wait_for_selector("button#mobile-film-picker", timeout=60000)
                except Exception as err:
                    print(f"No se encontró el botón de selector de película después de recargar: {err}")
                    continue

                await seleccionar_pelicula(page, peli)
                estado = await descargar_excel(page, peli)
                estado_peliculas.append({
                    "Pelicula": peli,
                    "Estado": estado,
                    "Fecha_batch": datetime.date.today().isoformat()
                })
                await cerrar_exports(page)

            except Exception as e:
                print(f"Error en {peli}: {e}")
                await forzar_reset_ui(page)
                print(f"🔄 Recargando página tras error en '{peli}'...")
                try:
                    await page.goto(url_reporte, wait_until="networkidle")
                    await page.wait_for_selector("button#mobile-film-picker", timeout=60000)
                except Exception as err:
                    print(f" Error recargando página tras fallo en '{peli}': {err}")
                    continue

                estado_peliculas.append({
                    "Pelicula": peli,
                    "Estado": f"❌ Error: {str(e).splitlines()[0]}",
                    "Fecha_batch": datetime.date.today().isoformat()
                })

        await context.close()
        await browser.close()

    # Exportar resultados
    df_estado = pd.DataFrame(estado_peliculas)
    excel_path = os.path.join(download_dir, f"estado_descarga_peliculas_{datetime.date.today().isoformat()}.xlsx")
    df_estado.to_excel(excel_path, index=False, engine="openpyxl")
    print(f"\n📄 Estado de descarga exportado a:\n{excel_path}")
```
