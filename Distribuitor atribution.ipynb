```python
import pandas as pd
from rapidfuzz import process, fuzz
from datetime import datetime
import os

# === FunciÃ³n para limpiar y matchear por similitud ===
def obtener_match_limpio(texto, lista_posibles, umbral=80):
    def clean(s):
        return s.lower().replace("the ", "").replace("a ", "").replace("an ", "").strip()

    texto_limpio = clean(texto)
    mapping = {clean(t): t for t in lista_posibles}
    lista_limpia = list(mapping.keys())

    match_limpio, score, _ = process.extractOne(
        query=texto_limpio,
        choices=lista_limpia,
        scorer=fuzz.ratio
    )

    if score >= umbral:
        return mapping[match_limpio], score
    else:
        return None, score

# === Matching con bo_dist ===
bo_matches = df_left['titulo_corto'].apply(
    lambda x: pd.Series(obtener_match_limpio(x, bo_dist['TITLE_STR']), index=['titulo_match_bo', 'similitud_bo'])
)

# Matching con awareness_dist ===
awareness_matches = df_left['titulo_corto'].apply(
    lambda x: pd.Series(obtener_match_limpio(x, awareness_dist['TITLE_STR']), index=['titulo_match_awareness', 'similitud_awareness'])
)

# === Unimos los resultados al dataframe base ===
df_matched = pd.concat([df_left, bo_matches, awareness_matches], axis=1)

# === Merge con bo_dist ===
bo_dist_clean = bo_dist.drop_duplicates(subset='TITLE_STR')
df_matched = df_matched.merge(bo_dist_clean, how='left', left_on='titulo_match_bo', right_on='TITLE_STR')
df_matched = df_matched.rename(columns={'DIST_STR': 'DIST_STR_matched'})

# === Merge con awareness_dist ===
awareness_dist_clean = awareness_dist.drop_duplicates(subset='TITLE_STR')
df_matched = df_matched.merge(awareness_dist_clean, how='left', left_on='titulo_match_awareness', right_on='TITLE_STR', suffixes=('', '_awareness'))
df_matched = df_matched.rename(columns={'STUDIO_STR': 'STUDIO_STR_matched'})

# === Resultado final con columnas relevantes ===
resultado = df_matched[[
    'titulo_corto',
    'titulo_match_bo', 'similitud_bo', 'DIST_STR_matched',
    'titulo_match_awareness', 'similitud_awareness', 'STUDIO_STR_matched'
]]

# === Exportar si querÃ©s ===
fecha = datetime.now().strftime("%Y-%m-%d_%H-%M")
ruta = f"/Workspace/Users/rocio.ramos@disney.com/Descargas/fuzzy_match_result_{fecha}.xlsx"
resultado.to_excel(ruta, index=False)

print(f"\nâœ… Resultado exportado a: {ruta}")
```
